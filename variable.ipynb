# Syntax of variable
  variable = value

  # variable = object = identifier 
  # variable never start with digit but it can ends with digits
  # Variable case sensitive (If user create variable with lower case, user must call with lower case only)
  # Special characters not allowed with variables except '_' (underscore)
  # Keywords never declare as variable , keywords == reserve words [35 reserved keywords]

```python
a = 5
# a is variable name, 5 is value, = is assignment operator
print(a)  # This will output: 5
```
