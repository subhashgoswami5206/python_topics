# TYPE CASTING
  # we convert one data type to another data type 
  # in type casting we cannot pass 2 arguments

# float to int

```python
int(5.4)
```

```python
int(54.0)
```

# boolean to int

```python
int(True)
```

```python
int(False)
```

# complex to integer type casting - not possible type casting

```python
int(1+2j) 
```

# Number string to int conversion

```python
int('10')
```

```python
int('ten')  # Number string to int is possible but text string string to int is not possible.
```

# Other datatypes to float conversion

```python
float(True)
```

```python
float(5.0)
```

```python
float('seven') # Text string to float is not possible.
```

```python
float('7.5') # Number string to float is possible but text string string to float is not possible.
```

```python
float(1+2j) # complex data type is also not possible to convert to float.
```
